```python
from imblearn.combine import SMOTETomek
stomek = SMOTETomek (random_state=0)
x_stomek, y_stomek = stomek.fit_sample(X_train, y_train)
x_stomek = pd.DataFrame(x_stomek,columns = X_train.columns)
```

```python
from imblearn.over_sampling import SMOTE
sm = SMOTE(random_state=0)
x_smote, y_smote = sm.fit_sample(X_train, y_train)
x_smote = pd.DataFrame(x_smote,columns = X_train.columns)
```

```python
# Modeling
```

```python
from keras import Sequential
from keras.layers import Dense, Dropout
```

```python
classifier = Sequential()
#First Hidden Layer
classifier.add(Dense(20, activation='relu', kernel_initializer='random_normal', input_dim=len(X.columns)))
classifier.add(Dropout(0.4))
#Second  Hidden Layer
classifier.add(Dense(20, activation='relu', kernel_initializer='random_normal'))
classifier.add(Dropout(0.4))
#Output Layer
classifier.add(Dense(1, activation='sigmoid', kernel_initializer='random_normal'))
```

```python
#Compiling the neural network
classifier.compile(optimizer ='rmsprop',loss='binary_crossentropy', metrics =['accuracy'])
```

```python
#Fitting the data to the training dataset
classifier.fit(x_smote,y_smote, batch_size=10, epochs=30, verbose=1)
```
